```python
import numpy as np
from copy import copy
import time
from IPython import display
import re

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
def getij(option, board):
    sel = np.where(board == option)[0]
    emp = np.where(board == '_')[0]
    i_pos = np.random.choice(sel)
    j_pos = np.random.choice(emp)
    return i_pos, j_pos


def checkVals(brd):
    selx = np.where(brd == 'x')[0]
    selo = np.where(brd == 'o')[0]

    sel = 'x'
    if len(selo) == 1:
        sel = 'x'
    elif len(selx) == 1:
        sel = 'o'
    return len(selo) == 1 or len(selx) == 1, sel


def getboard(board_str):
    data = []
    for i in range(0, len(board_str)):
        if board_str[i] == 'x':
            data.append(-1)
        elif board_str[i] == 'o':
            data.append(1)
        else:
            data.append(0)
    return data
```

```python
def play():
    board_str = 'xxxx__oooo'
    board = np.array(list(board_str))

    x = np.array([])
    y = np.array([])
    data = np.array([])
    for k in range(15):
        print("board: " + board_str)

        if k == 0:
            i = np.random.randint(4)
            j = np.random.randint(5, 6)
        else:
            i, j = getij('x', board)

        print('x goes from {:d} -> {:d}'.format(i, j))
        board[i] = '_'
        board[j] = 'x'
        board_str = ''.join(board)
        print("board: " + board_str)
        board_x = getboard(board_str)

        if k != 0:
            match_oxo = re.search("oxo", board_str)
            if match_oxo is not None:
                matchpos = re.search("oxo", board_str).start()
                x_pos = matchpos + 2 // 2
                board[x_pos] = '_'
                board_str = ''.join(board)
                print("x at position {:d}; killed".format(x_pos))

            match_xox = re.search("xox", board_str)
            if match_xox is not None:
                matchpos = re.search("xox", board_str).start()
                o_pos = matchpos + 2 // 2
                board[o_pos] = '_'
                board_str = ''.join(board)
                print("o at position {:d}; killed".format(o_pos))

        i, j = getij('o', board)

        board[i] = '_'
        board[j] = 'o'

        print('o goes from {:d} -> {:d}'.format(i, j))
        board_str = ''.join(board)
        print("board: " + board_str)
        board_o = getboard(board_str)

        checks1 = checkVals(board)
        if checks1[0]:
            data = np.concatenate([board_x, board_o])
            if k == 0:
                x = data
            else:
                x = np.vstack((x, data))
                x = x.astype(int)
            print(" ")
            print(checks1[1] + " won")
            y = (1 if checks1[1] == 'o' else 0) * np.ones(x.shape[0])
            y = y.astype(int)
            return x, y

        match_xox = re.search("xox", board_str)
        match_oxo = re.search("oxo", board_str)

        flag = False

        if match_xox is not None:
            flag = True
            matchpos = re.search("xox", board_str).start()
            o_pos = matchpos + 2 // 2
            board[o_pos] = '_'
            board_str = ''.join(board)
            print("o at position {:d}; killed".format(o_pos))

        if flag == False and match_oxo is not None:
            matchpos = re.search("oxo", board_str).start()
            x_pos = matchpos + 2 // 2
            board[x_pos] = '_'
            board_str = ''.join(board)
            print("x at position {:d}; killed".format(x_pos))

        data = np.concatenate([board_x, board_o])

        if k == 0:
            x = data
        else:
            x = np.vstack((x, data))
            x = x.astype(int)

        checks = checkVals(board)
        if checks[0]:
            print(" ")
            print(checks[1] + " won")
            y = (1 if checks[1] == 'o' else 0) * np.ones(x.shape[0])
            y = y.astype(int)
            break
    print("Done :")
    # if x.shape[0] != y.shape[0]:
    #     sel = 'o'
    #     y = (1 if sel == 'o' else 0) * np.ones(x.shape[0])
    #     y = y.astype(int)
    return x, y
```

```python
x1 = np.ndarray((0, 20))

y = np.ndarray((0, 0))

for i in range(100):
    vals = play()
    # x1 = np.atleast_2d(vals[0])
    if vals[0].shape[0] != vals[1].shape[0]:
        continue
    # print("x1 shape before:")
    # print(x1.shape)
    x1 = np.vstack((x1, vals[0]))
    x1 = x1.astype(int)
    # print("x1 shape after:")
    # print(x1.shape)
    # y=np.ndarray((vals[1].shape[0],))
    y = np.append(y, vals[1])
    y = y.astype(int)
    # y = (1 if vals[1][0] == 1 else 0) * np.ones(x1.shape[0])
    print("y")
print("done with the game")

```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-10, hidden_layer_sizes=(40, 4), activation='tanh', \
                    random_state=2, max_iter=100000000, learning_rate_init=0.001)

X_train, X_test, y_train, y_test = train_test_split(
    x1, y, test_size=0.5, random_state=2)

clf.fit(X_train, y_train)
print("training score", clf.score(X_train, y_train))

expected = y_test
predicted = clf.predict(X_test)
print("validation score", clf.score(X_test, y_test))
```
